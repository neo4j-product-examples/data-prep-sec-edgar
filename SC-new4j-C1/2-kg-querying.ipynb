# Intro to Cypher

Cypher is Neo4j's graph query language that allows users to store and retrieve data from the graph database. It is a declarative language, meaning that the user specifies what they want to retrieve, but not how to retrieve it. Cypher is a relatively simple language to learn, and is very powerful for querying graph data.

## Imports

```python
from dotenv import load_dotenv
import os

from langchain.graphs import Neo4jGraph
```

## Set up Neo4j

```python
# Load from environment
load_dotenv('.env', override=True)
NEO4J_URI = os.getenv('NEO4J_URI')
NEO4J_USERNAME = os.getenv('NEO4J_USERNAME')
NEO4J_PASSWORD = os.getenv('NEO4J_PASSWORD')
NEO4J_DATABASE = os.getenv('NEO4J_DATABASE') or 'neo4j'
OPENAI_API_KEY = os.getenv('OPENAI_API_KEY')
```

```python
# Create a knowledge graph using Langchain's Neo4j integration.
# This will be used for direct querying of the knowledge graph. 
kg = Neo4jGraph(
    url=NEO4J_URI, username=NEO4J_USERNAME, password=NEO4J_PASSWORD, database=NEO4J_DATABASE
)
```

### Single node patterns



```python
# MATCH a single node pattern, 
# binding the matched nodes to the variable `n`
# RETURN a count of the matched nodes
cypher = """
  MATCH (n) 
  RETURN count(n)
  """

kg.query(cypher)
```

```python
# Find all nodes that have a label Movie...

# MATCH a single node pattern of nodes labeled `Movie`
# binding the matched nodes to the variable `n`
# RETURN a count of the matched nodes
cypher = """
  MATCH (n:Movie) 
  RETURN count(n)
  """

kg.query(cypher)
```

```python
# Find all nodes that have a label Person...

# MATCH a single node pattern of nodes labeled `Person`
# binding the matched nodes to the variable `n`
# RETURN a count of the matched nodes
cypher = """
  MATCH (n:Person) 
  RETURN count(n)
  """

kg.query(cypher)
```

```python
# Find a particular Person by name...

# MATCH a single node pattern of nodes labeled `Person`
# with a property `name` equal to `Tom Hanks`
# binding the matched nodes to the variable `n`
# RETURN the matched node
cypher = """
  MATCH (n:Person {name:"Tom Hanks"}) 
  RETURN n
  """

kg.query(cypher)
```

```python
# Find a particular Movie by name...

# MATCH a single node pattern of nodes labeled `Movie`
# with a property `title` equal to `Cloud Atlas`
# binding the matched nodes to the variable `n`
# RETURN the matched node
cypher = """
  MATCH (n:Movie {title:"Cloud Atlas"}) 
  RETURN n
  """

kg.query(cypher)
```

### Single node patterns with predicates

```python
# Find old Movies from the '90s...

# MATCH a single node pattern of nodes labeled `Movie`
# WHERE the `released` property is between 1990 and 2000
# RETURN the titles from those movies
cypher = """
  MATCH (nineties:Movie) 
  WHERE nineties.released >= 1990 
    AND nineties.released < 2000 
  RETURN nineties.title
  """

kg.query(cypher)
```

### Pattern matching with multiple nodes

```python
# Find some Movies along with Actors ...

# MATCH a two node pattern of nodes labeled `Person`
# related to nodes labeled `Movie` by `ACTED_IN` relationships
# RETURN pairs of actor names and movie titles
cypher = """
  MATCH (actor:Person)-[:ACTED_IN]->(movie:Movie) 
  RETURN actor.name, movie.title LIMIT 10
  """

kg.query(cypher)
```

```python
# Find Movies that Tom Hanks acted in...

# MATCH a two node pattern of nodes labeled `Person`
# where the `name` property is equal to `Tom Hanks`
# related to nodes labeled `Movie` by `ACTED_IN` relationships
# RETURN the titles from those movies
cypher = """
  MATCH (tom:Person {name: "Tom Hanks"})-[:ACTED_IN]->(tomHanksMovies:Movie) 
  RETURN tom.name,tomHanksMovies.title
  """

kg.query(cypher)
```

```python
# Find who directed the movie "Cloud Atlas"...

# MATCH a two node pattern of nodes labeled `Movie`
# where the `title` property is equal to `Cloud Atlas`
# related to nodes labeled `Person` by `DIRECTED` relationships
# RETURN the names of the directors
cypher = """
  MATCH (cloudAtlas:Movie {title: "Cloud Atlas"})<-[:DIRECTED]-(directors:Person) 
  RETURN directors.name
  """

kg.query(cypher)
```

#### Find Tom Hanks' co-actors

An important note on this one: the `ACTED_IN` relationship appears twice in the query, once from Tom to some movie
then again from that movie to some other actor. But Tom doesn't appear in the results. Why is that? It's because
relationships are only traversed once when matching a pattern.

```python
# Find actors who have been in the same movies as Tom Hanks, his co-actors...

# MATCH a three node pattern from nodes labeled `Person`
# where the `name` property is equal to `Tom Hanks`
# through the `ACTED_IN` relationships to a movies bound to `m`
# then from those movies back through `ACTED_IN` relationships 
# to `coActors` nodes
# RETURN the names of the co-actors and the movie they were in together with Tom
cypher = """
  MATCH (tom:Person {name:"Tom Hanks"})-[:ACTED_IN]->(m)<-[:ACTED_IN]-(coActors) 
  RETURN coActors.name, m.title
  """

kg.query(cypher)
```

## Vector search with Cypher



```python
## Create a vector index for movie descriptions
kg.query("""
  CREATE VECTOR INDEX movie_tagline_embeddings IF NOT EXISTS
  FOR (m:Movie) ON (m.taglineEmbedding) 
  OPTIONS { indexConfig: {
    `vector.dimensions`: 1536,
    `vector.similarity_function`: 'cosine'
  }}"""
)
```

```python
# Add embeddings to the vector index by calling to OpenAI API directly from Cypher
# 
# MATCH a single node pattern of nodes labeled `Movie`
# WITH those movies and a function call to encode the tagline property using OpenAI
# CALL db.create.setNodeVectorProperty to set the `taglineEmbedding` property on the movie node 
kg.query("""
    MATCH (movie:Movie) WHERE movie.tagline IS NOT NULL
    WITH movie, genai.vector.encode(movie.tagline, "OpenAI", {token: $openAiApiKey}) AS vector
    CALL db.create.setNodeVectorProperty(movie, "taglineEmbedding", vector)
    """, 
    params={"openAiApiKey":OPENAI_API_KEY} )
```

```python
question = "What movies are about love?"
kg.query("""
    WITH genai.vector.encode($question, "OpenAI", {token: $openAiApiKey}) AS embedding
    CALL db.index.vector.queryNodes('movie_tagline_embeddings', $top_k, embedding) YIELD node AS movie, score
    RETURN movie.title + " -- " + movie.tagline AS result
    """, 
    params={"openAiApiKey":OPENAI_API_KEY,
            "question": question,
            "top_k": 5
            })
```
